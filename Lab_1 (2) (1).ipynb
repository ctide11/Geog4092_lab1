```python
import pandas as pd

merge1 = pd.merge(poly1, table1, left_on='MUSYM', right_on='musym')
merge2 = pd.merge(poly2, table2, left_on='MUSYM', right_on='musym')
merge3 = pd.merge(poly3, table3, left_on='MUSYM', right_on='musym')
merge4 = pd.merge(poly4, table4, left_on='MUSYM', right_on='musym')
merge5 = pd.merge(poly5, table5, left_on='MUSYM', right_on='musym')
merge6 = pd.merge(poly6, table6, left_on='MUSYM', right_on='musym')
merge7 = pd.merge(poly7, table7, left_on='MUSYM', right_on='musym')
merge8 = pd.merge(poly8, table8, left_on='MUSYM', right_on='musym')
merge9 = pd.merge(poly9, table9, left_on='MUSYM', right_on='musym')
```

```python
polygons = []

tables = []

for poly_name in layers:
    if poly_name.startswith('soilmu'):
        feature_class = gpd.read_file('./lab1.gpkg', layer=poly_name)
        polygons.append(feature_class)
       
       
        
for tabl_name in layers:
    if tabl_name.startswith('muaggatt'):
        feature_2 = gpd.read_file('./lab1.gpkg', layer=tabl_name)
        tables.append(feature_2)
                            
```

```python
# .assign lets you add a new field to an existing data set.
# mapunit is the new field and ='...' is what youre adding into the field.

merge1 = merge1.assign(mapunit = 'co618')
merge2 = merge2.assign(mapunit = 'co001')
merge3 = merge3.assign(mapunit = 'co641')
merge4 = merge4.assign(mapunit = 'co642')
merge5 = merge5.assign(mapunit = 'co643')
merge6 = merge6.assign(mapunit = 'co644')
merge7 = merge7.assign(mapunit = 'co645')
merge8 = merge8.assign(mapunit = 'co651')
merge9 = merge9.assign(mapunit = 'co653')
```

```python
# pd.concat will merge all 9 feature classes into 1 feature class.

MergeAll = pd.concat([merge1, merge2, merge3, merge4, merge5, merge6, merge7, merge8, merge9])

MergeAll.head()
```

```python
# .set_geometry will select column youre wanting to plot "geometry_x"

MergeAll = MergeAll.set_geometry('geometry_x')
```

```python
# To insersect watershed data with feature class, you must first read/create a dataset..
# for the watershed data like you did with the other feature classes.

watershed = gpd.read_file('./lab1.gpkg', layer='wbdhu8_lab1')

# To intersect both data sets, use gpd.overlay function

intersect = gpd.overlay(watershed, MergeAll, how='intersection')
```

```python
# To find number of features that correspond to each watersshed.

total_features = intersect.groupby(by=['NAME']).count()
```

```python
# .loc lets you access data values in a particular row or column.
# f' allows you to use the values found using .loc when printing

big_thom = total_features.loc['Big Thompson', 'STATES']
st_v = total_features.loc['St. Vrain', 'STATES']

print(f'The total number of features intersected at Big Thompson is {big_thom}')
print(f'The total number of features intersected at St. Vrain is {st_v}')
intersect.plot()
```
